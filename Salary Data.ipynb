Analysis: 

From the Data, we see that Majority of the Employees in the Company have Barchelor's Degree holder as their highest degree with 3023 people which is about 45 percent of the total number of Employees. Followed by Master's Degree Holders who are 1861 people (about 27.7%) in Number with the least employed qualification being High School Degree (6.68%) and PhD Degree holders accounting for 20.62%. Among the Employees, three Genders were Represented (Male, Female and Others). Among the Genders represented Male were the most employed, making up about 54.8 percent (3674 out of 6702) of the employees while Women occupy about 45% with 0.2 percent identifying as Others.

The Highest Paid Salary was 250,000 and the lowest Paid Salary is 350 and the Average Salary given to the staffs in the company is 115326.9. The Highest Paid Male Employees were 3 in number and they were all earning 250,000 while the highest Paid Female was paid 220,000 and the highest paid Non Binary (Others) were paid 166,109. The Lowest Salary earned by the male Gender is 350 while that of the female Gender is  500 while that of those who identify as others is 62,852. However when we we check the average salary across each Gender, we see that those who Identify as others edge it slightly as they earn about 125869.86 compared to 121389.87 and  107888.99 for the male and Female Genders respectively. 

The Average Salary by Educational Level as shown in Output 121 above tells us that salary increases with educational Level. Also, we see a positive correlation between Years of Experience and the Educatonal Level of the Employees (i.e Those with Higher Educational Qualification have more years of Experience than those with Lower Educational Level/Qualification). We see that 89% (2,696 out of 3,023) of Barchelor's Degree Holders have Years of Experience between 0 - 10 years and 97.32 percent (436 out of 448) of high School Degree holders have Years Of Experience of between 0 - 10 years. When we compare to the for Master's and PhD degree holders where just 61.8 percent and 22.1 percent have below 10 years Experience.

Finally, we see that Salary tend to increase across the age groups and by how long you stay at the company when we look at the average Salary by Age and average Salary by Years of Experience. And when we take a look at our comparison between age and Educational level, we see that similarly to our last comparison between Years of Experience and Educational Level, Most of the younger demographics have the lower qualifications as their highest form of education. We see that most Barchelor degree holders in the company are aged between 20 and 35 (about 85.77%) and which is true also for those who are high school degree holders (97.7%). However for the PhD and Master's Degree holders we see that 71.37% and 41.27% of their respective holders are above 35 years.
